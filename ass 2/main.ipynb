```python
import torch
import torch.nn as nn  
from utils.ResNet import CIFAR_ResNet
print(torch.__version__)
print(torch.cuda.is_available())
```

```python
pureNet = CIFAR_ResNet(method = "pure",
                    lr = 1e-2, 
                    lr_decay = 0.95,
                    loss_func = nn.CrossEntropyLoss(), 
                    weight_decay = 1e-4, 
                    epochs = 25, 
                    denote_path = "./train_history/pure",
                    verbose = True)
pureNet.train()
pureNet.save("pureNet")
```

```python
cutoutNet = CIFAR_ResNet(method = "cutout",
                    lr = 1e-2, 
                    lr_decay = 0.95,
                    loss_func = nn.CrossEntropyLoss(), 
                    weight_decay = 1e-4, 
                    epochs = 25, 
                    denote_path = "./train_history/cutout",
                    verbose = True)
cutoutNet.train()
cutoutNet.save("cutoutNet")
```

```python
mixupNet = CIFAR_ResNet(method = "mixup",
                    lr = 1e-2, 
                    lr_decay = 0.95,
                    loss_func = nn.CrossEntropyLoss(), 
                    weight_decay = 1e-4, 
                    epochs = 25, 
                    denote_path = "./train_history/mixup",
                    verbose = True)
mixupNet.train()
mixupNet.save("mixupNet")
```

```python
cutmixNet = CIFAR_ResNet(method = "cutmix",
                    lr = 1e-2, 
                    lr_decay = 0.95,
                    loss_func = nn.CrossEntropyLoss(), 
                    weight_decay = 1e-4, 
                    epochs = 25, 
                    denote_path = "./train_history/cutmix",
                    verbose = True)
cutmixNet.train()
cutmixNet.save("cutmixNet")
```

```python
import matplotlib.pyplot as plt
plt.title('Test accuracy using different methods')
plt.plot(pureNet.val_accuracy_list, '-o', label='pure ResNet')
plt.plot(cutoutNet.val_accuracy_list, '-o', label='ResNet with cutout data')
plt.plot(mixupNet.val_accuracy_list, '-o', label='ResNet with mixup data')
plt.plot(cutmixNet.val_accuracy_list, '-o', label='ResNet with cutmix data')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
```
